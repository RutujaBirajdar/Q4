# Q4. Polynomial Plotting with Annotations:

Randomly select the coefficients of a 5th degree polynomial. Set the random seed as the last two 
digits of your roll number. Plot the polynomial for -5 ≤ x ≤ 5. Annotate the plot to identify the 
maxima and minima of the polynomial.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(8)

coeffs = np.random.randn(6)
def poly(x, c):
    return np.polyval(c, x)
```

```python
x_vals = np.linspace(-5, 5, 1000)

y_vals = poly(x_vals, coeffs)

max_x = x_vals[np.argmax(y_vals)]
max_y = np.max(y_vals)
min_x = x_vals[np.argmin(y_vals)]
min_y = np.min(y_vals)
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Polynomial')

plt.scatter([max_x, min_x], [max_y, min_y], color='red')
plt.annotate('Maxima', xy=(max_x, max_y), xytext=(-4, 150), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Minima', xy=(min_x, min_y), xytext=(2, -300), arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel('x')
plt.ylabel('y')
plt.title('5th Degree Polynomial with Maxima and Minima')

plt.legend()
plt.grid(True)
plt.show()
```
